```python
def delete_buku(df, judul, penulis, genre, penerbit):
    # Check if the book with the given attributes exists
    mask = (df['Judul'] == judul) & (df['Penulis'] == penulis) & (df['Genre'] == genre) & (df['Penerbit'] == penerbit)
    
    if df.loc[mask].empty:
        print("Book not found. No book was deleted.")
    else:
        # Remove the rows that match the specified conditions
        df = df.loc[~mask].reset_index(drop=True)
        print("The book has been deleted.")
    
    return df

# Example usage to delete a specific book
Buku = delete_buku(Buku, judul='Game of Hearts: Love In Vegas', penulis='Silvarani', genre='fiksi_sastra',
                   penerbit='Gramedia Pustaka Utama')

# Print the updated 'Buku' DataFrame after deleting the book
print(Buku)
```
